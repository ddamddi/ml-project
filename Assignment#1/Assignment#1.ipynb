```python
import torch
import matplotlib
import matplotlib.pyplot as plt
import numpy as np

%matplotlib inline

NUMOFDATA = 1000
```

```python
# SET CENTROID POINT
c0 = np.array([3,4]).reshape(2,1)
c1 = np.array([1,2]).reshape(2,1)

# CREATING DATASET FOR TRAINING AND TESTING
train_true = c0 + np.random.randn(2, int(NUMOFDATA/2))
train_false = c1 + np.random.randn(2, int(NUMOFDATA/2))
test_true = c0 + np.random.randn(2, int(NUMOFDATA/2))
test_false = c1 + np.random.randn(2, int(NUMOFDATA/2))

o = np.ones((1, int(NUMOFDATA/2)))
z = np.zeros((1, int(NUMOFDATA/2)))

Y = np.concatenate((o,z), axis=1)
```

```python

fig, ax = plt.subplots(1,3,figsize=(18,5))
# fig.suptitle("DATA SET", fontsize='20')

ax[0].set_title("Training Set")
ax[0].plot(train_true[0], train_true[1],'bo', label='True')
ax[0].plot(train_false[0], train_false[1],'ro', label='False')
ax[0].grid(True)
ax[0].legend()

ax[1].set_title("Testing Set")
ax[1].plot(test_true[0], test_true[1],'go', label='True')
ax[1].plot(test_false[0], test_false[1],'yo', label='False')
ax[1].grid(True)
ax[1].legend()

ax[2].set_title("Superimposed Set")
ax[2].plot(train_true[0], train_true[1],'bo', label='Train True')
ax[2].plot(test_true[0], test_true[1],'go', label='Test True')
ax[2].plot(train_false[0], train_false[1],'ro', label='Train False')
ax[2].plot(test_false[0], test_false[1],'yo', label='Test False')
ax[2].grid(True)
ax[2].legend()

plt.show()
```

```python
LEARNING_RATE = 0.05
ITERATIONS = 500

b = 0
w = np.zeros((2,1))
X_train = np.concatenate((train_true,train_false), axis=1)
X_test = np.concatenate((test_true,test_false), axis=1)

train_loss = np.zeros((2,ITERATIONS))
train_accuracy = np.zeros((2,ITERATIONS))
test_loss = np.zeros((2,ITERATIONS))
test_accuracy = np.zeros((2,ITERATIONS))

for i in range(ITERATIONS):
    J = 0
    dw = np.zeros((2,1))
    db = 0
    
    z = np.dot(w.T, X_train) + b   # Y = WX + b
    A = 1/(1 + np.exp(-z))        # APPLY SIGMOID FUNCTION
    
    # APPLY GRADIENT DESCENT ALGORITHM
    dz = A-Y
    dw = np.dot(X_train,dz.T)/NUMOFDATA
    db = np.sum(dz)/NUMOFDATA
    
    w = w - LEARNING_RATE*dw
    b = b - LEARNING_RATE*db
    
    # FOR CALCULATING TRAIN LOSS
    J = (np.dot(Y, (np.log(A)).T) + np.dot(1-Y, (np.log(1-A)).T))
    J = -np.sum(J)/NUMOFDATA
    train_loss[0][i] = i+1
    train_loss[1][i] = J
    
    # FOR TRAIN ACCURACY
    for x in range(NUMOFDATA):
        if A[0][x] >= 0.5:
            A[0][x] = 1
        else:
            A[0][x] = 0
    
    correct = 0
    for x in range(NUMOFDATA):
        if A[0][x] == Y[0][x]:
            correct += 1
        
    train_accuracy[0][i] = i+1
    train_accuracy[1][i] = correct/NUMOFDATA
    
    # FOR CALCULATING TEST LOSS
    z = np.dot(w.T, X_test) + b
    A = 1/(1 + np.exp(-z))
    
    J = (np.dot(Y, (np.log(A)).T) + np.dot(1-Y, (np.log(1-A)).T))
    J = -np.sum(J)/NUMOFDATA
    test_loss[0][i] = i+1
    test_loss[1][i] = J
    
    # FOR TEST ACCURACY
    for x in range(NUMOFDATA):
        if A[0][x] >= 0.5:
            A[0][x] = 1
        else:
            A[0][x] = 0
    
    correct = 0
    for x in range(NUMOFDATA):
        if A[0][x] == Y[0][x]:
            correct += 1
        
    test_accuracy[0][i] = i+1
    test_accuracy[1][i] = correct/NUMOFDATA
```

```python
# PLOT TRAINING AND TESTING LOSS AT EVERY ITERATION
fig, ax2 = plt.subplots(1,2,figsize=(20,6))
ax2[0].set_title("Loss Value")
ax2[0].set_ylabel("Loss", fontsize="12")
ax2[0].set_xlabel("iteration", fontsize="12")
ax2[0].plot(train_loss[0], train_loss[1],'-r', label='Train Loss')
ax2[0].plot(test_loss[0], test_loss[1], '-b', label='Test Loss')
ax2[0].legend(fontsize="12")

# PLOT TRAINING AND TESTING ACCURACY AT EVERY ITERATION
ax2[1].set_title("Accuracy")
ax2[1].set_ylabel("Accuracy", fontsize="12")
ax2[1].set_xlabel("iteration", fontsize="12")
ax2[1].plot(train_accuracy[0], train_accuracy[1], '-y', label='Train Accuracy')
ax2[1].plot(test_accuracy[0], test_accuracy[1], '-g', label='Test Accuracy')
ax2[1].legend(fontsize="12")
plt.show()
```
